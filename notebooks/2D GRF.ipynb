```python
rshape = (256,256)
deltas=1
nk = 12
nangle=1
Pk = power.PowSpec(power.gaussian)
xx = np.arange(rshape[0])
yy = np.arange(rshape[1])
X, Y = np.meshgrid(xx, yy)
```

```python
karr = np.arange(0,10,0.1)
plt.plot(karr, Pk(karr, sigma_k=1))
```

```python
field_k = realization.dft_realizn(rshape, Pk=lambda k:Pk(k, sigma_k=0.05))
field = np.fft.irfft2(field_k)
```

```python
%matplotlib inline
plt.imshow(field)
```

```python
#%matplotlib osx 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y, field)
```

```python
#%matplotlib osx 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,Y, field)
fig.show()
```
